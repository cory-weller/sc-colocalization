```python
import pandas as pd
```

```python
def extract_rsid(df):
    rsid_list = []
    for snp in df['TopSNP']:
        if ':' in snp:
            rsid_list.append(snp.split(':')[2])
        else:
            rsid_list.append(snp)
        
    df['SNP'] = rsid_list
    return df
```

```python
# read in formatted summary statistics - gr38
ad_sum = pd.read_csv('/data/CARD/projects/singlecell_humanbrain/data/final_formatted_sumstats/AD_Bellenguez.formatted.tsv', sep = '\t')
lbd_sum = pd.read_csv('/data/CARD/projects/singlecell_humanbrain/data/final_formatted_sumstats/LBD_Chia.formatted.tsv', sep = '\t')
pd_all_sum = pd.read_csv("/data/CARD/projects/singlecell_humanbrain/data/final_formatted_sumstats/PD_Nalls.formatted.tsv", sep = '\t')
```

```python
# read in eQTL data - gr38
BrainCortex_eqtl = pd.read_csv('/data/CARD/OTHER/nick_scratch/MAGMA/Coloc_files/Cortex-EUR-out.cis.summary.txt', sep = '\t')
Basalganglia_eqtl = pd.read_csv('/data/CARD/OTHER/nick_scratch/MAGMA/Coloc_files/Basalganglia-EUR-out.cis.summary.txt', sep = '\t')
Hippo_eqtl = pd.read_csv('/data/CARD/OTHER/nick_scratch/MAGMA/Coloc_files/Hippocampus-out.cis.summary.txt', sep = '\t')
Spinal_eqtl = pd.read_csv('/data/CARD/OTHER/nick_scratch/MAGMA/Coloc_files/Spinalcord-out.cis.summary.txt', sep = '\t')
Cerebellum_eqtl = pd.read_csv('/data/CARD/OTHER/nick_scratch/MAGMA/Coloc_files/Cerebellum-out.cis.summary.txt', sep = '\t')
```

```python
# reduce down summary statistic dfs to only significant gwas hits
ad_hits = ad_sum.query('P < 5e-8')
lbd_hits = lbd_sum.query('P < 5e-8')
pd_hits = pd_all_sum.query('P < 5e-8')
```

```python
ad_hits['varbeta'] = ad_hits.loc['SE'] * ad_hits.loc['SE']
lbd_hits['varbeta'] = lbd_hits.loc['SE'] * lbd_hits.loc['SE']
pd_hits['varbeta'] = pd_hits.loc['SE'] * pd_hits.loc['SE']
```

```python
# add column with RSID
BrainCortex_eQTL = extract_rsid(BrainCortex_eqtl)
Basalganglia_eqtl = extract_rsid(Basalganglia_eqtl)
Hippo_eqtl = extract_rsid(Hippo_eqtl)
Spinal_eqtl = extract_rsid(Spinal_eqtl)
Cerebellum_eqtl = extract_rsid(Cerebellum_eqtl)
```

```python
# Check beta equals 0
print(sum(BrainCortex_eQTL.b == 0))
print(sum(Basalganglia_eqtl.b == 0))
print(sum(Hippo_eqtl.b == 0))
print(sum(Spinal_eqtl.b == 0))
print(sum(Cerebellum_eqtl.b == 0))
```

```python
BrainCortex_eQTL.rename(columns={"b": "b_MA"})
Basalganglia_eqtl.rename(columns={"b": "b_MA"})
Hippo_eqtl.rename(columns={"b": "b_MA"})
Spinal_eqtl.rename(columns={"b": "b_MA"})
Cerebellum_eqtl.rename(columns={"b": "b_MA"})
```

```python
# find SNPs shared between GWAS sum stats and eQTL summary stats
ad_cortex = ad_sum.merge(BrainCortex_eQTL, on = 'SNP')
ad_bg = ad_sum.merge(Basalganglia_eqtl, on = 'SNP')
ad_hippo = ad_sum.merge(Hippo_eqtl, on = 'SNP')
ad_spinal = ad_sum.merge(Spinal_eqtl, on = 'SNP')
ad_cerebellum = ad_sum.merge(Cerebellum_eqtl, on = 'SNP')
```
